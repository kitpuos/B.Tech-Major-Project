```python
def hash_trick(hash_value, parts):
    """
    Maps a hash value to a list of integers of fixed length.
    """
    # Convert the hash value to bytes
    hash_bytes = hash_value.encode('utf-8')

    # Assuming hash_bytes is your hash value in bytes
    binary_string = bin(int(hash_bytes.hex(), 16))[2:]

    # Split the binary string into parts based on the number of parts
    part_length = len(binary_string) // parts
    hash_parts = [binary_string[i * part_length: (i + 1) * part_length] for i in range(parts)]

    # Convert each part to an integer
    int_values = [int(part, 2) for part in hash_parts]

    # Reduce the integer value to a manageable range
    prime_modulus = 1009  # Choose a large prime number
    rem = [(num % prime_modulus) for num in int_values]

    return rem
```

```python
# Apply the function to your DataFrame to create a new column with a list of integers
df['md5_int_list'] = df['md5'].apply(lambda x: hash_trick(x, 4))
df['sha1_int_list'] = df['sha1'].apply(lambda x: hash_trick(x, 5))
df['sha256_int_list'] = df['sha256'].apply(lambda x: hash_trick(x, 8))
df
```

```python
# Expand the lists into multiple columns
df[['md5_int1', 'md5_int2', 'md5_int3', 'md5_int4']] = pd.DataFrame(df['md5_int_list'].tolist(), index=df.index)
df[['sha1_int1', 'sha1_int2', 'sha1_int3', 'sha1_int4', 'sha1_int5']] = pd.DataFrame(df['sha1_int_list'].tolist(), index=df.index)
df[['sha256_int1', 'sha256_int2', 'sha256_int3', 'sha256_int4', 'sha256_int5', 'sha256_int6', 'sha256_int7', 'sha256_int8']] = pd.DataFrame(df['sha256_int_list'].tolist(), index=df.index)
df
```

```python
# Drop the intermediate columns
df.drop(['id', 'md5', 'sha1', 'sha256', 'total', 'positives', 'filetype', 'submitted', 'user_id',
         'md5_int_list', 'sha1_int_list', 'sha256_int_list'], axis=1, inplace=True)
df
```
